## Control

```python
msg = 'No se cumplio ninguna condicion'

for num in lista:
    if num % 2 == 0:
        print('Primera condicion: ', num)
    elif num % 3 == 0:
        print('Segunda condicion: ', num)
    else:
        print(msg)
```

## Compresion de listas vs bucles for 

```python

#Esta es una list comprehension que itera sobre cada elemento "num" en la lista "lista"
#y divide cada elemento entre 2, creando una nueva lista con los resultados
[num/2 for num in lista]

#Esta es una list comprehension que itera sobre cada elemento "num" en la lista "lista"
#y divide cada elemento entre 2 si "num" es par, creando una nueva lista con los resultados
[num/2 for num in lista if num % 2 == 0]

#Esta es una lista vacía que se usará para almacenar los resultados
l = []

#Esta es una estructura de iteración "for" que itera sobre cada elemento "num" en la lista "lista"
for num in lista:

    #Si "num" es par, se divide entre 2 y se añade el resultado a la lista "l"
    if num%2 == 0:
        l.append(num/2)

#Esta es una list comprehension que itera sobre cada elemento "num" en la lista "lista"
#y divide cada elemento entre 2 si "num" es igual a 2, sino se añade el elemento tal cual a la lista
[num/2 if num == 2 else num for num in lista]

#Esta es una lista vacía que se usará para almacenar los resultados
l = []

#Esta es una estructura de iteración "for" que itera sobre cada elemento "num" en la lista "lista"
for num in lista:

    #Si "num" es par, se divide entre 2 y se añade el resultado a la lista "l"
    #Si "num" es impar, se añade el elemento tal cual a la lista "l"
    if num%2 == 0:
        l.append(num/2)
    else:
        l.append(num)
```

## modificacion de valores 

```python

df = pd.read_csv('/workspaces/CH-32845/data/raw/data_kaggle.csv')
df_copy = df.copy() 

df_copy['Car Parks'].fillna(0,inplace=True)

df_copy[['Car Parks']].fillna(0,inplace=True)
```

```python
df_copy['Bathrooms'][:10] = 0
# https://pandas.pydata.org/pandas-docs/stable/user_guide/indexing.html#returning-a-view-versus-a-copy 
```

```python
### Ambas operaciones son equivalentes
### No generan warnings porque se esta modificando el objeto entero 
### En este caso una serie 
df_copy['Car Parks'] = df_copy['Car Parks'].fillna(0)
df_copy['Car Parks'].fillna(0,inplace=True)

# Genera un warning porque se esta modificando una parte del objeto 
# En este caso una columna de un dataframe 
df_copy[['Car Parks']].fillna(0,inplace=True) 

## No genera un warning porque se esta modificando el objeto entero 
df_copy.fillna(0,inplace=True)
```
