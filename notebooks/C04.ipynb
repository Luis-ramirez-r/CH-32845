```python
# Utilizar time delta para calcular la diferencia entre fechas 
# Reemplazar los valores faltantes en las fechas 
# ====== 
```

```python
# presidentes_archivo
```

```python
inicio_fin = presidentes_archivo[['start','end']][0:40].copy()
```

```python
# convertir a datetime 
inicio_fin['start'] = pd.to_datetime(inicio_fin['start'],dayfirst=True)
inicio_fin['end'] = pd.to_datetime(inicio_fin['end'],dayfirst=True)
# add a column for the delta 
inicio_fin['delta'] = inicio_fin['end'] - inicio_fin['start']
# get the mode to the delta 
mode_time = inicio_fin['delta'].mode()[0]

mode_time
```

```python
# sume the mode to the start date in the presidentes_archivo
presidentes_archivo['end_app'] = presidentes_archivo['start'] + mode_time
```

```python
df_lluvias  = pd.read_csv(csv_files[2])
df_lluvias.head()
```

```python
df_lluvias = df_lluvias.set_index('Year')
```

```python
df_lluvias.sum(axis='columns').to_frame().T
```

```python
suma_lluvia = df_lluvias.sum(axis='columns')
```

```python
#
plt.plot(df_lluvias.index,df_lluvias.sum(axis='columns'))
```

```python
fig, ax = plt.subplots(figsize=(6,3))
ax.plot(suma_lluvia.index,suma_lluvia.values,label='Lluvias') 
ax.set_title('Lluvias ')
ax.set_xlabel('AÃ±o')
ax.set_ylabel('Lluvias (mm)')
ax.grid(True)
ax.legend()
plt.show()
```

```python
fig.savefig('../data/reports/lluvias.png')
```
